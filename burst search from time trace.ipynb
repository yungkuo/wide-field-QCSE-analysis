# Functions

```python
def movingaverage(interval, window_size, mode='valid'):
    window = np.ones(int(window_size))/float(window_size)
    return np.convolve(interval, window, mode)
```

```python
def get_dF(timetrace, frame_start=2):
    assert frame_start%2 == 0
    res = np.diff(timetrace[frame_start:])
    res[::2] *= -1
    return res
```

```python
def get_dFF(timetrace, frame_start=2, running_avg_F=False, stitchends=True, window_size=4):
    assert frame_start%2 == 0
    assert window_size%2 == 0
    dF = np.diff(timetrace[frame_start:])
    dF[::2] *= -1
    F = timetrace[frame_start+1::2].repeat(2)[:-1]
    dFF = dF/F
    if running_avg_F:
        if not stitchends:
            F = movingaverage(F, window_size)
            dFF = dF[window_size//2:-window_size//2+1]/F
        else:
            F = timetrace[frame_start+1::2].repeat(2)
            F = np.append(np.append(F[-window_size//2:],F), F[:window_size//2])
            F = movingaverage(F, window_size)[:len(dF)]
            dFF = dF/F
    return dFF, F
```

```python
MovingAvg = pd.DataFrame()
for count in L:
    ttf = F[count]
    ttl = L[count]
    ttw = W[count]
    _min = ttl.median()-30
    _max = ttl.median()+30
    for a, l in enumerate(ttl[frame_start:]):
        if np.logical_or(l>_max, l<_min):
            ttl[a+frame_start] = ttl[a+frame_start-1]
    dll, l = get_dFF(ttl, frame_start)

    dff, f = get_dFF(ttf, frame_start)
    _min = np.median(df)-5
    _max = np.median(df)+5
    dff[np.logical_or(df>_max, df<_min)] = 0
    
    dfwhm, fwhm = get_dFF(ttw, frame_start)
    _min = 0
    _max = np.mean(ttw)+80
    fwhm[np.logical_or(fwhm>_max, fwhm<_min)] = 0
    
    scoref = movingaverage(dff, m)
    scorel = movingaverage(dll*l, m)
    MvAvgl = movingaverage(l, m)
    MvAvgf = movingaverage(f, m)
    MvAvgw = movingaverage(fwhm, m)
    
    DF1 = pd.DataFrame()
    DF1['dL'] = scorel
    DF1['L'] = MvAvgl
    DF1['dF/F'] = scoref
    DF1['F'] = MvAvgf
    DF1['FWHM'] = MvAvgw
    MovingAvg = MovingAvg.append(DF1)
```
